```python
from tkinter import *
import requests
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
class App:

    def __init__(self, master):

        frame = Frame(master)
      
        #create navigator label and place it in grid
        navigator_label = Label(root, text = "Navigator password")
        navigator_label.grid(row=0,column = 0)
        
        #create navigator entry and place it.
        navigator_entry = Entry(root,show=" ")
        navigator_entry.grid(row=0,column = 1, sticky = E)
        
        #create city label and place it.
        city_label = Label(root, text = "CI/TY password")
        city_label.grid(row = 1, column = 0)
        
        #create city entry and place it.
        city_pw = Entry(root,show =" ")
        city_pw.grid(row=1, column = 1) 


        #create buttons to quit app
        self.quit = Button( root, text="QUIT", fg="red", command=frame.quit )
        self.quit.grid(row =4, column = 0)
        
        #create fetch button function that goes to navigator
        def fetch_navigator(event):
            driver = webdriver.Chrome('c:/chromedriver')  # Optional argument, if not specified will search path.
            driver.get('https://psavfs.psav.com/adfs/ls/?wa=wsignin1.0&wtrealm=http%3a%2f%2fapps.psav.com%2fnavigator&wctx=rm%3d0%26id%3dpassive%26ru%3d%252f&wct=2015-05-27T18%3a05%3a25Z');
            time.sleep(1) # Let the user actually see something!

            #finds username and puts it in
            search_UN = driver.find_element_by_name('UserName')
            search_UN.send_keys('bodonnell@psav.com')

            #finds password box and puts it in
            searchPW = driver.find_element_by_name('Password')
            searchPW.send_keys( navigator_entry.get() )

            submit_box =driver.find_element_by_id('submitButton')
            submit_box.submit()

            driver.implicitly_wait(30)

            searchOrder = driver.find_element_by_id("searchOrder")
            searchOrder.send_keys(Keys.RETURN)
            print("completed", navigator_entry)

        #create fetch button
        button_get_navigator = Button(root, text = "Open Navigator")
        button_get_navigator.grid(row = 0, column = 3)
        button_get_navigator.bind("<Button-1>", fetch_navigator)
        
        #create function to fetch CITY
        def fetch_city(event):
            driver = webdriver.Chrome('c:/chromedriver')
            driver.get('https://salesnet.marriott.com/gsemr5_enu/start.swe?SWECmd=GotoView&SWEView=TNT+CMT+Site+FSS+Diary+-+My+Property&SWERF=1&SWEHo=salesnet.marriott.com&SWEBU=1&SWEApplet0=TNT+SHM+Site+List+Applet+-+1+Row&SWERowId0=M-TIEG')
        
            #enter username password
            search_UN = driver.find_element_by_name('username')
            search_UN.send_keys('bodon128')
            search_pw = driver.find_element_by_name('password')
            search_pw.send_keys(city_pw.get())
            submit_box = driver.find_element_by_class_name('btn')
            submit_box.submit()
            
        #create fetch button for CITY
        button_get_city = Button(root, text = "Open CI/TY")
        button_get_city.grid(row = 1, column = 3, sticky = W)
        button_get_city.bind("<Button-1>",fetch_city)


        
      
        
                
        
root = Tk()

app = App(root)
root.geometry("400x300")
root.mainloop()
root.destroy() # optional; see description below
```
